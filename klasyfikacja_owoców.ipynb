```python
import warnings
import os
import subprocess
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, BatchNormalization, Dropout
from tensorflow.keras.mixed_precision import set_global_policy
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from collections import Counter

# Wyłącz ostrzeżenia specyficzne dla Keras
warnings.filterwarnings("ignore", category=UserWarning, module="keras.src.trainers.data_adapters.py_dataset_adapter")
warnings.filterwarnings("ignore", category=UserWarning, module="keras.src.trainers.epoch_iterator")

# Wyłącz wszystkie ostrzeżenia i komunikaty informacyjne TensorFlow
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# Definiujemy adres URL zbioru danych i ścieżki
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/4yIRGlIpNfKEGJYMhZV52g/fruits-360-original-size.zip"
local_zip = "fruits-360-original-size.zip"
extract_dir = "fruits-360-original-size"

def download_dataset(url, output_file):
    """Pobiera zbiór danych przy użyciu wget w trybie cichym (quiet mode)."""
    print("Pobieranie zbioru danych...")
    subprocess.run(["wget", "-q", "-O", output_file, url], check=True)  # `-q` dla trybu cichego
    print("Pobieranie zakończone.")

def extract_zip_in_chunks(zip_file, extract_to, batch_size=2000):
    """
    Rozpakowuje duży plik ZIP w partiach, aby uniknąć problemów z pamięcią.
    Przetwarza określoną liczbę plików (batch_size) jednorazowo.
    """
    print("Rozpakowywanie zbioru danych w partiach...")
    os.makedirs(extract_to, exist_ok=True)  # Upewnij się, że katalog docelowy istnieje
    
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        files = zip_ref.namelist()  # Lista wszystkich plików w archiwum
        total_files = len(files)
        
        for i in range(0, total_files, batch_size):
            batch = files[i:i+batch_size]
            for file in batch:
                zip_ref.extract(file, extract_to)  # Rozpakuj każdy plik w partii
            print(f"Rozpakowano {min(i+batch_size, total_files)} z {total_files} plików...")
    
    print(f"Zbiór danych został pomyślnie rozpakowany do '{extract_to}'.")

def initial_script():
    # Główne uruchomienie skryptu
    if __name__ == "__main__":
        # Pobierz zbiór danych, jeśli jeszcze nie został pobrany
        if not os.path.exists(local_zip):
            download_dataset(url, local_zip)
        else:
            print("Zbiór danych został już pobrany.")
        
        # Rozpakuj zbiór danych, jeśli jeszcze nie został rozpakowany
        if not os.path.exists(extract_dir):
            extract_zip_in_chunks(local_zip, extract_dir)
        else:
            print("Zbiór danych został już rozpakowany.")
        
        # (Opcjonalnie) usuń plik zip
        if os.path.exists(local_zip):
            os.remove(local_zip)
            print(f"Usunięto plik zip: {local_zip}")

#initial_script()

# Inicjalizacja liczników dla klas rzeczywistych i przewidywanych
actual_count = Counter()
predicted_count = Counter()

# Funkcja zwracająca nazwę klasy na podstawie przewidzianego indeksu
def get_class_name_from_index(predicted_index, class_index_mapping):
    """Konwertuje przewidziany indeks na nazwę klasy."""
    for class_name, index in class_index_mapping.items():
        if index == predicted_index:
            return class_name
    return "Unknown"

# Funkcja do wizualizacji obrazu wraz z przewidywaną i rzeczywistą klasą
def visualize_prediction_with_actual(img_path, class_index_mapping):
    # Pobierz nazwę prawidłowej klasy na podstawie ścieżki do pliku (nazwa folderu)
    class_name = os.path.basename(os.path.dirname(img_path))

    # Wczytaj i przetwórz obraz
    img = load_img(img_path, target_size=(64, 64))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Przewidywanie klasy
    prediction = model.predict(img_array)
    predicted_index = np.argmax(prediction, axis=-1)[0]
    predicted_class_name = get_class_name_from_index(predicted_index, class_index_mapping)

    # Aktualizacja liczników
    actual_count[class_name] += 1
    predicted_count[predicted_class_name] += 1

    # Wizualizacja obrazu z podpisami
    plt.figure(figsize=(2, 2), dpi=100)
    plt.imshow(img)
    plt.title(f"Rzeczywista klasa: {class_name}, Przewidziana klasa: {predicted_class_name}")
    plt.axis('off')
    plt.show()

# Ścieżki do zbiorów danych
train_dir = 'fruits-360-original-size/fruits-360-original-size/Training'
val_dir = 'fruits-360-original-size/fruits-360-original-size/Validation'
test_dir = 'fruits-360-original-size/fruits-360-original-size/Test'

# Generatory danych obrazowych
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Wczytanie obrazów z katalogów
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),
    batch_size=16,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(64, 64),
    batch_size=16,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64, 64),
    batch_size=16,
    class_mode='categorical'
)

# Wczytanie modelu VGG16 z wagami wstępnie wytrenowanymi
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(64, 64, 3))

# Zamrożenie warstw modelu bazowego
for layer in base_model.layers:
    layer.trainable = False

# Budowa modelu
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(train_generator.num_classes, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# Definicja callbacków
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Włączenie mieszanej precyzji (jeśli używamy GPU)
set_global_policy('float32')

steps_per_epoch = 50
validation_steps = 25

# Trenowanie modelu
history = model.fit(
    train_generator,
    epochs=5,
    validation_data=val_generator,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    callbacks=[lr_scheduler, early_stopping]
)

# Sprawdzenie liczby warstw w modelu bazowym
num_layers = len(base_model.layers)
print(f"Model bazowy ma {num_layers} warstw.")

# Odmrożenie ostatnich 5 warstw w celu dalszego dostrajania (fine-tuning)
for layer in base_model.layers[-5:]:
    layer.trainable = True

# Zamrożenie warstw BatchNormalization w celu przyspieszenia fine-tuningu
for layer in base_model.layers:
    if isinstance(layer, tf.keras.layers.BatchNormalization):
        layer.trainable = False

# Rekompilacja modelu z szybszym optymalizatorem
model.compile(
    loss='categorical_crossentropy',
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    metrics=['accuracy']
)

# Kontynuacja trenowania z mniejszą liczbą kroków na epokę
history_fine = model.fit(
    train_generator,
    epochs=5,
    validation_data=val_generator,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    callbacks=[lr_scheduler, early_stopping]
)

# Ewaluacja na zbiorze testowym
test_loss, test_accuracy = model.evaluate(test_generator, steps=50)
print(f"Dokładność testu: {test_accuracy:.2f}")

# Wykres dokładności
plt.plot(history.history['accuracy'], label='Dokładność trenowania')
plt.plot(history.history['val_accuracy'], label='Dokładność walidacji')
plt.plot(history_fine.history['accuracy'], label='Dokładność trenowania (fine-tuning)')
plt.plot(history_fine.history['val_accuracy'], label='Dokładność walidacji (fine-tuning)')
plt.xlabel('Epoki')
plt.ylabel('Dokładność')
plt.legend()
plt.title('Dokładność trenowania i walidacji')
plt.grid(True)
plt.show()

# Wykres straty (loss)
plt.plot(history.history['loss'], label='Strata trenowania')
plt.plot(history.history['val_loss'], label='Strata walidacji')
plt.plot(history_fine.history['loss'], label='Strata trenowania (fine-tuning)')
plt.plot(history_fine.history['val_loss'], label='Strata walidacji (fine-tuning)')
plt.xlabel('Epoki')
plt.ylabel('Strata')
plt.legend()
plt.title('Strata trenowania i walidacji')
plt.grid(True)
plt.show()

# Pobranie mapowania indeksów klas z generatora treningowego
class_index_mapping = train_generator.class_indices
print("Mapowanie indeksów klas:", class_index_mapping)

# Przykładowe obrazy do testu
sample_images = [
    'fruits-360-original-size/fruits-360-original-size/Test/apple_braeburn_1/r0_11.jpg',
    'fruits-360-original-size/fruits-360-original-size/Test/pear_1/r0_103.jpg',
    'fruits-360-original-size/fruits-360-original-size/Test/cucumber_3/r0_103.jpg',
]

# Uruchomienie predykcji i wizualizacji
for img_path in sample_images:
    visualize_prediction_with_actual(img_path, class_index_mapping)
```
