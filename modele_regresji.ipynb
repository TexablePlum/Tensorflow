```python
import os
# Wyłączenie optymalizacji OneDNN dla TensorFlow oraz ustawienie poziomu logów
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import pandas as pd
import numpy as np
import keras
import warnings

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input

# Tłumienie ostrzeżeń typu FutureWarning (budowa czystego środowiska)
warnings.simplefilter('ignore', FutureWarning)

# Budowa modelu regresyjnego (definicja struktury sieci)
def regression_model():
    # Utworzenie modelu sekwencyjnego
    model = Sequential()
    # Dodanie warstwy wejściowej (budowa wejścia) o kształcie odpowiadającym liczbie kolumn
    model.add(Input(shape=(n_cols,)))
    # Dodanie pierwszej warstwy ukrytej z 50 neuronami i funkcją aktywacji ReLU
    model.add(Dense(50, activation='relu'))
    # Dodanie drugiej warstwy ukrytej z 50 neuronami i funkcją aktywacji ReLU
    model.add(Dense(50, activation='relu'))
    # Dodanie warstwy wyjściowej (budowa wyjścia) z 1 neuronem
    model.add(Dense(1))
    
    # Kompilacja modelu: używamy optymalizatora Adam i funkcji straty - średni błąd kwadratowy
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Pobranie zbioru danych o betonie (budowa danych)
filepath = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
concrete_data = pd.read_csv(filepath)

# Podgląd pierwszych rekordów
concrete_data.head()

# Sprawdzenie wymiarów zbioru danych
concrete_data.shape

# Podstawowe statystyki zbioru danych
concrete_data.describe()

# Sprawdzenie brakujących wartości
concrete_data.isnull().sum()

# Wyodrębnienie nazw kolumn
concrete_data_columns = concrete_data.columns

# Budowa zbioru predyktorów i zmiennej docelowej:
# - predictors: wszystkie kolumny poza 'Strength'
# - target: kolumna 'Strength'
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']

# Podgląd zbioru predyktorów
predictors.head()

# Podgląd zmiennej docelowej
target.head()

# Normalizacja predyktorów: standaryzacja (budowa standaryzacji)
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()

# Ustalenie liczby predyktorów (budowa ustawienia parametrów wejściowych)
n_cols = predictors_norm.shape[1]

# Budowa modelu regresji
model = regression_model()

# Dopasowanie modelu do danych:
# - validation_split: procent danych użyty do walidacji
# - epochs: liczba epok treningowych
# - verbose: poziom szczegółowości logów treningu
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```
