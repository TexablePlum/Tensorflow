```python
import os
import numpy as np
import warnings
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization, Reshape
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'

# Funkcja budująca model generatora
def build_generator():
    # Model sekwencyjny Keras
    model = Sequential()
    # Pierwsza warstwa gęsta (ang. Dense), przyjmuje 100 liczb losowych (wejście - wektor szumu)
    model.add(Dense(256, input_dim=100))
    # Aktywacja LeakyReLU (działa jak ReLU, ale dopuszcza niewielkie ujemne wartości)
    model.add(LeakyReLU(alpha=0.2))
    # BatchNormalization stabilizuje proces uczenia (normalizacja danych w mini-batchach)
    model.add(BatchNormalization(momentum=0.8))

    # Kolejna warstwa
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization(momentum=0.8))

    # Następna warstwa
    model.add(Dense(1024))
    model.add(LeakyReLU(alpha=0.2))
    model.add(BatchNormalization(momentum=0.8))

    # Ostatnia warstwa zmienia wymiar na 28*28*1, bo obrazy w MNIST mają rozmiar 28x28 i 1 kanał (skala szarości)
    # Używamy aktywacji 'tanh', która zwraca wartości z zakresu [-1, 1]
    model.add(Dense(28 * 28 * 1, activation='tanh'))
    # Reshape do postaci obrazu 28x28x1
    model.add(Reshape((28, 28, 1)))

    return model


# Funkcja budująca model dyskryminatora
def build_discriminator():
    model = Sequential()
    # Najpierw "spłaszczamy" obraz 28x28x1 do wektora
    model.add(Flatten(input_shape=(28, 28, 1)))
    # Warstwa gęsta
    model.add(Dense(512))
    model.add(LeakyReLU(alpha=0.2))

    # Kolejna warstwa gęsta
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.2))

    # Ostatecznie wyjście 1 neuron z aktywacją sigmoid, który mówi nam, czy obraz jest prawdziwy (1) czy fałszywy (0)
    model.add(Dense(1, activation='sigmoid'))

    return model

# Funkcja tworząca pełny model GAN, łącząc generator i dyskryminator
def build_gan(generator, discriminator):
    # Zamrażamy wagi dyskryminatora przy trenowaniu GAN (aby przy kroku uczenia generatora nie zmieniać dyskryminatora)
    discriminator.trainable = False

    # Definiujemy wejście do GAN (szum, wektor 100 liczb)
    gan_input = Input(shape=(100,))
    # Przepuszczamy szum przez generator
    generated_image = generator(gan_input)
    # Następnie gotowy obraz przez dyskryminator
    gan_output = discriminator(generated_image)

    # Budujemy model, który łączy oba etapy
    gan = Model(gan_input, gan_output)
    # Kompilujemy używając straty binarnej crossentropii i optymalizatora 'adam'
    gan.compile(loss='binary_crossentropy', optimizer='adam')

    return gan

# Funkcja do generowania i wyświetlania przykładowych obrazów w czasie treningu
def sample_images(generator, epoch, num_images=25):
    # Generujemy losowy szum (wektor) z rozkładu normalnego o średniej 0 i odchyleniu 1
    noise = np.random.normal(0, 1, (num_images, 100))
    # Przepuszczamy szum przez generator, aby otrzymać obrazy
    generated_images = generator.predict(noise)
    # Przeskalowanie pikseli z zakresu [-1,1] na [0,1] (wygodne do wyświetlania)
    generated_images = 0.5 * generated_images + 0.5

    # Tworzymy siatkę 5x5, aby wyświetlić 25 obrazów
    fig, axs = plt.subplots(5, 5, figsize=(10, 10))
    count = 0

    for i in range(5):
        for j in range(5):
            axs[i, j].imshow(generated_images[count, :, :, 0], cmap='gray')
            axs[i, j].axis('off')
            count += 1
    plt.show()

# Wyłączamy ostrzeżenia Pythona (np. deprecations itp.)
warnings.filterwarnings('ignore')

# Ładujemy zbiór MNIST
(x_train, _), (_, _) = mnist.load_data()

# Normalizujemy piksele do zakresu [-1, 1]
x_train = x_train.astype('float32') / 127.5 - 1.
# Dodajemy dodatkowy wymiar, aby otrzymać kształt (liczba_próbek, 28, 28, 1)
x_train = np.expand_dims(x_train, axis=-1)

# Sprawdźmy kształt danych
print(x_train.shape)

# Budujemy generator
generator = build_generator()
generator.summary()

# Budujemy i kompilujemy dyskryminator
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
discriminator.summary()

# Budujemy GAN łącząc generator i dyskryminator
gan = build_gan(generator, discriminator)
gan.summary()

# Parametry treningu
batch_size = 64
epochs = 50
sample_interval = 10  # co ile epok wyświetlamy próbki wygenerowanych obrazów

# Przygotowujemy etykiety - wektory 1 (obraz prawdziwy) i 0 (obraz fałszywy)
real = np.ones((batch_size, 1))
fake = np.zeros((batch_size, 1))

# Główna pętla treningowa
for epoch in range(epochs):
    # 1. Trenujemy dyskryminator
    # Losowo wybieramy batch (batch_size) indeksów z treningowego zbioru MNIST
    idx = np.random.randint(0, x_train.shape[0], batch_size)
    # Pobieramy prawdziwe obrazy
    real_images = x_train[idx]
    # Generujemy szum i przepuszczamy go przez generator (obrazy fałszywe)
    noise = np.random.normal(0, 1, (batch_size, 100))
    generated_images = generator.predict(noise)

    # Trenujemy dyskryminator na prawdziwych obrazach
    d_loss_real = discriminator.train_on_batch(real_images, real)
    # Trenujemy dyskryminator na fałszywych obrazach
    d_loss_fake = discriminator.train_on_batch(generated_images, fake)
    # Średnia strata dyskryminatora
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # 2. Trenujemy generator (poprzez GAN)
    # Ponownie generujemy szum
    noise = np.random.normal(0, 1, (batch_size, 100))
    # Tutaj w trenowaniu całego GAN jako etykiety używamy 'real' - generator stara się "oszukać" dyskryminator,
    # żeby myślał, że generowane obrazy są prawdziwe
    g_loss = gan.train_on_batch(noise, real)

    # Wyświetlamy stan co określoną liczbę epok
    if epoch % sample_interval == 0:
        print(f"{epoch} [D loss: {d_loss[0]}] [D accuracy: {100 * d_loss[1]}%] [G loss: {g_loss}]")

# Po skończonym treningu wyświetlamy przykładowe obrazy wygenerowane przez generator
sample_images(generator, epochs)

# Na koniec obliczamy dokładność dyskryminatora na próbkach prawdziwych i fałszywych
noise = np.random.normal(0, 1, (batch_size, 100))
generated_images = generator.predict(noise)

# Ocena dyskryminatora na prawdziwych obrazach
real_images = x_train[np.random.randint(0, x_train.shape[0], batch_size)]
d_loss_real = discriminator.evaluate(real_images, np.ones((batch_size, 1)), verbose=0)

# Ocena dyskryminatora na fałszywych obrazach
d_loss_fake = discriminator.evaluate(generated_images, np.zeros((batch_size, 1)), verbose=0)

print(f"Discriminator Accuracy on Real Images: {d_loss_real[1] * 100:.2f}%")
print(f"Discriminator Accuracy on Fake Images: {d_loss_fake[1] * 100:.2f}%")
```
