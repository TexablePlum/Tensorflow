```python
import numpy as np
import matplotlib.pyplot as plt

# Budowa funkcji aktywacji Sigmoid oraz jej pochodnej
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

# Budowa funkcji aktywacji ReLU oraz jej pochodnej
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return np.where(z > 0, 1, 0)

# Budowa funkcji aktywacji tanh oraz jej pochodnej
def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z) ** 2

# Generowanie zakresu wartoci wejciowych (budowa danych wejciowych)
z = np.linspace(-10, 10, 400)

# Obliczenie gradientw funkcji aktywacyjnych Sigmoid i ReLU
sigmoid_grad = sigmoid_derivative(z)
relu_grad = relu_derivative(z)

# Budowa wykresu funkcji aktywacyjnych Sigmoid i ReLU wraz z ich gradientami
plt.figure(figsize=(12, 6))

# Wykres Sigmoid oraz jej pochodnej
plt.subplot(1, 2, 1)
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label="Sigmoid Derivative", color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

# Wykres ReLU oraz jej pochodnej
plt.subplot(1, 2, 2)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()


# Generowanie nowego zakresu wartoci wejciowych (budowa danych wejciowych)
z = np.linspace(-5, 5, 100)

# Pobranie gradientw dla funkcji ReLU i tanh
tanh_grad = tanh_derivative(z)
relu_grad = relu_derivative(z)

# Budowa kolejnego wykresu funkcji aktywacyjnych ReLU i tanh wraz z ich gradientami
plt.figure(figsize=(12, 6))

# Wykres ReLU oraz jej pochodnej
plt.subplot(1, 2, 1)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

# Wykres tanh oraz jej pochodnej
plt.subplot(1, 2, 2)
plt.plot(z, tanh(z), label='tanH Activation', color='g')
plt.plot(z, tanh_grad, label="tanH Derivative", color='r', linestyle='--')
plt.title('tanH Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
